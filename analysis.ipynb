# Fun Club Impact Analytics
Analyzing student participation, volunteer hours, and program growth.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/fun_club_monthly_program_data.csv')
df.head()
```

```python
program_summary = df.groupby('Program Area')[['Students Participated','Instructional Sessions','Volunteer Hours']].sum()
program_summary
```

```python
monthly = df.groupby('Month')['Students Participated'].sum().reset_index()
monthly.plot(x='Month', y='Students Participated', figsize=(14,6), marker='o', title='Monthly Student Participation Trend')
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()
```
